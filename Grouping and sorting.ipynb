# Grouping and sorting


```python
import pandas as pd
import numpy as np
pd.set_option('max_rows', 5)
```

```python

reviews = pd.read_csv("oscar_age_female.csv")
pd.set_option('max_rows', 5)
reviews.head()
```

```python
reviews.groupby('Age').count()
```

```python
countries_reviewed = reviews.groupby('Age').Movie.agg([len])
countries_reviewed.head()
```

```python
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```
